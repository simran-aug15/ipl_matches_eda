#### Missing values

```python
# Which columns contains missing values, and how many missing values does each column have?
df.isnull().sum()
```

```python
# What percentage of missing values is present in each column?
missing_value=(df.isnull().sum()/len(df))*100
print(missing_value[missing_value>0])
```

```python
# Decide how to handle missing values(drop,fill with mode/median,or keep)?
df.select_dtypes(include="object").isnull().sum()
```

```python
df.select_dtypes(include="number").isnull().sum()
```

```python
# Fill categorical columns with mode
for i in df.select_dtypes(include="object"):
    df[i]=df[i].fillna(df[i].mode()[0])   # To get the actual value otherwise return series 
```

```python
df.head()
```

```python
# Fill numerical columns with median
for i in df.select_dtypes(include="number").columns:
    df[i]=df[i].fillna(df[i].median()) 
```

```python
df.isnull().sum()
```

#### Univariate Analysis

```python
df.head()
```

```python
# Plot the distribution of match winners.Which team has won the most matches?
df["WinningTeam"].value_counts()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(data=df,x="WinningTeam",order=df["WinningTeam"].value_counts().index)  #order=sort bars (highest to lowest),index=gets team name after sorting
plt.xticks(rotation=90)
plt.title("Distribution of Match Winner")
```

```python
# Plot the count of matches played in each season using countplot or barplot?
a=df["Season"].value_counts()
a
```

```python
plt.figure(figsize=(15,8))
sns.countplot(data=df,x="Season",order=df["Season"].value_counts().index)
plt.title("Number of matches Played in Each season")
```

#### Bivariate Analysis

```python
df.head()
```

```python
# Compare the number of matches won by each team when playing at different venues?
venue_win=df.groupby("Venue")["WinningTeam"].value_counts()
venue_win
```

```python
plt.figure(figsize=(15,8))
sns.countplot(data=df,x="Venue",hue="WinningTeam")
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyze the relationship between Toss Winner and Match Winner?
df.groupby("TossWinner")["WinningTeam"].value_counts()
```

```python
#check True or Flase
df["Toss_match_winner"]=df["TossWinner"]==df["WinningTeam"]
df["Toss_match_winner"].value_counts()
```

```python
sns.countplot(data=df,x="Toss_match_winner")
plt.title("Did Toss winning leads to match winning")
```

```python
# What is the most frequent categories in each categorical column?
cat_col=df.select_dtypes(include=["category","object"])
cat_col
for i in cat_col:
    counts=df[i].value_counts()
    print(counts.head(20))
    plt.figure(figsize=(8,3))
    sns.countplot(data=df,x=i,order=counts.index[:20])
    plt.xticks(rotation=90)
```

```python
# Compare team1 and team2
sns.violinplot(x="Team1",y="Team2",data=df)
plt.title("Violin plot: Team1 vs Team2")
plt.xticks(rotation=90)
```
